# Voting records
### Predict depending on some answers which party one person might agree with

```python
# Necessary imports
import seaborn as sns
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt

# Import the data set into a DataFrame
data = pd.read_csv('votes.csv')
data
```

#### Plot the votes of both parties on education policy acceptance
The resulting plot shows the difference in voting behavior between the two parties for the 'education' bill, with each party colored differently.

It seems like Democrats voted resoundingly against this bill, compared to Republicans. 

```python
plt.figure()
sns.countplot(x='education',hue='party',data=data)
plt.xticks([0,1],['No','Yes'])
plt.show()
```

#### Counter plot for satellite policy

```python
sns.countplot(x='satellite',data=data,hue='party')
plt.xticks([0,1],['No','Yes'])
plt.show()
```

#### Counter plot for missile policy

```python
sns.countplot(x='missile',data=data,hue='party')
plt.xticks([0,1],['No','Yes'])
plt.show()
```
