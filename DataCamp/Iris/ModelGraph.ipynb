## Plot Model graph for various K values
We will plot the train and test accuracy for multiple values of k (0-9) 

```python
# Imports
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt

# Values of K
K = np.arange(1,9)

# Accuracies
train_acc = np.empty(len(K)) 
test_acc = np.empty(len(K)) 

# import the Iris dataset
iris = datasets.load_iris()

# Split the train and test data
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.3,
                                                 random_state=21,stratify=iris.target)

# Iterate over values of k
for i,k in enumerate(K):
    # Set up a KNN classifier 
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Train the model
    knn.fit(x_train,y_train)
    
    # Note down the accuracies
    test_acc[i] = knn.score(x_test,y_test)
    train_acc[i] = knn.score(x_train,y_train)
```

### Plot the accuracy comparison graph

```python
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(K,test_acc,label = "Test Accuracy")
plt.plot(K,train_acc,label = "Train Accuracy")
plt.legend()
plt.xlabel("No of neighbors")
plt.ylabel('Accuracy')
plt.show()
```
