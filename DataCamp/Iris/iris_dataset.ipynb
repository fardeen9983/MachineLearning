# Iris Dataset

```python
# Import dataset from sklearn
from sklearn import  datasets

# Other imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set plotting style to ggplot
plt.style.use('ggplot')

# Load iris data set
iris = datasets.load_iris()

# Checkout the type of the iris variable
type(iris)
```

```python
# Bunch is similar to a dictionary key value pairs

# ALl the keys in the dataset
print("Keys : ",iris.keys())

# Shape of the database
print("Shape : ",iris.data.shape)

# Data and target type - numpy array 
print("Types : ",type(iris.data),type(iris.target))

# Target names and indexes
iris.target_names
```

## Exploratory Data Analysis

```python
# Create a data frame from iris data and target
x = iris.data
y = iris.target

data = pd.DataFrame(x,columns=iris.feature_names)
data
```

## Visual EDA

```python
from pandas.plotting import scatter_matrix

# Produce a scatter matrix using pandas 
scatter_matrix(data,c = y,figsize=[12,12],s=150,marker='D')
```

### Scatter matrix
It plots all possible graphs by varying the axises from the given data's feature set  
Arguments
* arg0 - Dataframe to plot
* c - the target dataframe by which we will color code our data points
* figsize - Physical dimensions of the plot size
* s - size of the points plotted opn the graph
